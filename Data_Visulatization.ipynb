```python
# ===============================================
# üìä Mental Health and Social Media Visualization
# ===============================================

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Step 2: Upload and Load Dataset
from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Mental_Health_and_Social_Media_Balance_Dataset.csv")

# Step 3: Basic Data Overview
print("üîπ Dataset Shape:", df.shape)
print("\nüîπ Columns:", df.columns.tolist())
display(df.head())
display(df.describe())

# Step 4: Data Info & Missing Values
df.info()
print("\nMissing Values:\n", df.isnull().sum())

# ------------------------------------------------
# üî∏ Basic Data Analysis Visualizations
# ------------------------------------------------

# 1Ô∏è‚É£ Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Gender', palette='coolwarm')
plt.title("Gender Distribution")
plt.show()

# 2Ô∏è‚É£ Social Media Platform Usage
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Social_Media_Platform', palette='Set2')
plt.title("Most Used Social Media Platform")
plt.xticks(rotation=45)
plt.show()

# 3Ô∏è‚É£ Histogram - Daily Screen Time
plt.figure(figsize=(7,5))
plt.hist(df['Daily_Screen_Time(hrs)'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Daily Screen Time (hours)")
plt.xlabel("Daily Screen Time (hrs)")
plt.ylabel("Frequency")
plt.show()

# 4Ô∏è‚É£ Line Chart - Average Stress Level vs Age
age_stress = df.groupby('Age')['Stress_Level(1-10)'].mean().reset_index()
plt.figure(figsize=(8,5))
plt.plot(age_stress['Age'], age_stress['Stress_Level(1-10)'], color='purple')
plt.title("Average Stress Level vs Age")
plt.xlabel("Age")
plt.ylabel("Mean Stress Level (1-10)")
plt.show()

# 5Ô∏è‚É£ Box Plot - Screen Time by Gender
plt.figure(figsize=(7,5))
sns.boxplot(data=df, x='Gender', y='Daily_Screen_Time(hrs)', palette='Pastel1')
plt.title("Screen Time by Gender")
plt.show()

# 6Ô∏è‚É£ Violin Plot - Happiness by Social Media Platform
plt.figure(figsize=(8,5))
sns.violinplot(data=df, x='Social_Media_Platform', y='Happiness_Index(1-10)', palette='muted')
plt.title("Happiness Distribution by Platform")
plt.xticks(rotation=45)
plt.show()

# 7Ô∏è‚É£ Count Plot - Exercise Frequency Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Exercise_Frequency(week)', color='teal')
plt.title("Exercise Frequency per Week")
plt.show()

# ------------------------------------------------
# üî∏ Plotly Interactive Charts
# ------------------------------------------------

# Interactive Scatter Plot: Screen Time vs Happiness
fig1 = px.scatter(
    df,
    x='Daily_Screen_Time(hrs)',
    y='Happiness_Index(1-10)',
    color='Gender',
    size='Exercise_Frequency(week)',
    hover_data=['Social_Media_Platform'],
    title="Interactive Scatter: Screen Time vs Happiness Index"
)
fig1.show()

# Interactive Histogram: Sleep Quality Distribution
fig2 = px.histogram(
    df,
    x='Sleep_Quality(1-10)',
    color='Social_Media_Platform',
    nbins=20,
    title="Interactive Histogram: Sleep Quality by Platform"
)
fig2.show()

# ------------------------------------------------
# üî∏ Correlation Heatmap
# ------------------------------------------------
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap - Mental Health Metrics")
plt.show()
```
